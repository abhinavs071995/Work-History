```python
import pandas as pd

# Ensure necessary date columns are in datetime format
df_lead['lead_created_at'] = pd.to_datetime(df_lead['lead_created_at'])
df_lead['first_status_except_created_or_contacted_date'] = pd.to_datetime(df_lead['first_status_except_created_or_contacted_date'])
df_task['task_created_at'] = pd.to_datetime(df_task['task_created_at'])

# Merge df_task with df_lead to bring in the relevant dates
merged_tasks = pd.merge(df_task, df_lead[['lead_id', 'lead_created_at', 'first_status_except_created_or_contacted_date']], on='lead_id')

# Filter for tasks created within the specified date range
filtered_tasks = merged_tasks[(merged_tasks['task_created_at'] >= merged_tasks['lead_created_at']) & 
                              (merged_tasks['task_created_at'] <= merged_tasks['first_status_except_created_or_contacted_date'])]

# Count the total tasks and violated tasks per lead_id
tasks_count_per_lead = filtered_tasks.groupby('lead_id').size().reset_index(name='total_tasks')
violated_tasks_per_lead = filtered_tasks[filtered_tasks['is_violated'] == 1].groupby('lead_id').size().reset_index(name='violated_tasks')

# Merge the task counts back into df_lead, using a left join to keep all rows in df_lead
df_lead = pd.merge(df_lead, tasks_count_per_lead, on='lead_id', how='left')
df_lead = pd.merge(df_lead, violated_tasks_per_lead, on='lead_id', how='left')

# Fill NaN values with 0 for leads with no tasks or no violated tasks within the range
df_lead['total_tasks'].fillna(0, inplace=True)
df_lead['violated_tasks'].fillna(0, inplace=True)

# Convert counts to integers
df_lead['total_tasks'] = df_lead['total_tasks'].astype(int)
df_lead['violated_tasks'] = df_lead['violated_tasks'].astype(int)
```

```python
import pandas as pd

# Ensure the date columns are in datetime format for accurate comparison
df_lead['lead_created_at'] = pd.to_datetime(df_lead['lead_created_at'])
df_lead['first_status_except_created_or_contacted_date'] = pd.to_datetime(df_lead['first_status_except_created_or_contacted_date'])
df_note['note_created_at'] = pd.to_datetime(df_note['note_created_at'])

# Merge df_note with df_lead to bring in the relevant date columns for each lead
merged_notes = pd.merge(df_note, df_lead[['lead_id', 'lead_created_at', 'first_status_except_created_or_contacted_date']], on='lead_id')

# Filter notes created within the specified date range for each lead
filtered_notes = merged_notes[(merged_notes['note_created_at'] >= merged_notes['lead_created_at']) & 
                              (merged_notes['note_created_at'] <= merged_notes['first_status_except_created_or_contacted_date'])]

# Count the total notes per lead_id within the specified range
notes_count_per_lead = filtered_notes.groupby('lead_id').size().reset_index(name='total_notes')

# Merge the notes counts back into df_lead, using a left join to ensure all rows in df_lead are preserved
df_lead = pd.merge(df_lead, notes_count_per_lead, on='lead_id', how='left')

# Fill NaN values with 0 for leads with no notes in the specified range
df_lead['total_notes'].fillna(0, inplace=True)

# Convert 'total_notes' to integer, as counts should be whole numbers
df_lead['total_notes'] = df_lead['total_notes'].astype(int)
```

```python
# Ensure datetime columns are in the correct format
df_lead['lead_created_at'] = pd.to_datetime(df_lead['lead_created_at'])
df_note['note_created_at'] = pd.to_datetime(df_note['note_created_at'])
df_task['task_created_at'] = pd.to_datetime(df_task['task_created_at'])

# Merge df_lead with df_note and df_task to align lead creation times
notes_with_lead_time = pd.merge(df_lead[['lead_id', 'lead_created_at']], df_note, on='lead_id')
tasks_with_lead_time = pd.merge(df_lead[['lead_id', 'lead_created_at']], df_task, on='lead_id')

# Filter notes and tasks that were created after the lead's creation time
filtered_notes = notes_with_lead_time[notes_with_lead_time['note_created_at'] > notes_with_lead_time['lead_created_at']]
filtered_tasks = tasks_with_lead_time[tasks_with_lead_time['task_created_at'] > tasks_with_lead_time['lead_created_at']]

# Group by lead_id and find the earliest note and task created after lead_created_at
first_note_per_lead = filtered_notes.groupby('lead_id')['note_created_at'].min().reset_index(name='first_note_created_at')
first_task_per_lead = filtered_tasks.groupby('lead_id')['task_created_at'].min().reset_index(name='first_task_created_at')

# Merge these minimum times back into df_lead
df_lead = pd.merge(df_lead, first_note_per_lead, on='lead_id', how='left')
df_lead = pd.merge(df_lead, first_task_per_lead, on='lead_id', how='left')
```

```python
# Calculate the time difference in hours from lead_created_at to first_note_created_at
df_lead['hours_to_first_note'] = (df_lead['first_note_created_at'] - df_lead['lead_created_at']).dt.total_seconds() / 3600

# Calculate the time difference in hours from lead_created_at to first_task_created_at
df_lead['hours_to_first_task'] = (df_lead['first_task_created_at'] - df_lead['lead_created_at']).dt.total_seconds() / 3600
```

```python
import numpy as np

# Define the bins and labels for the time to first note
bins = [-np.inf, 0, 2, 6, 24, 48, 72, np.inf]
labels = ['No Notes', 'Less than 2 hours', '2 hours to 6 hours', '6 hours to 24 hours', 
          '24 hours to 48 hours', '48 hours to 72 hours', 'More than 72 hours']

# Create a new column for the time to first note bucket
# NaN values in 'hours_to_first_note' indicate no notes, so we fill these with -1 before binning
df_lead['time_to_first_notes_bucket'] = pd.cut(df_lead['hours_to_first_note'].fillna(-1), bins=bins, labels=labels, right=False)
```

```python
# Calculate the task violation percentage
df_lead['task_violation_percentage'] = (df_lead['violated_tasks'] / df_lead['total_tasks']) * 100

# Define a function to categorize the task violation percentage into buckets
def task_violation_bucket(row):
    if row['total_tasks'] == 0:
        return 'No tasks assigned'
    elif row['task_violation_percentage'] == 0:
        return 'Zero task violation'
    elif row['task_violation_percentage'] < 20 and row['task_violation_percentage'] > 0:
        return 'Less than 20% task violation'
    elif 20 <= row['task_violation_percentage'] <= 50:
        return '20%-50% task violation'
    else:
        return 'Greater than 50%'

# Apply the function to create a new column for the task violation percentage bucket
df_lead['task_violation_percentage_bucket'] = df_lead.apply(task_violation_bucket, axis=1)
```

```python
# Convert 'qualified_date' to datetime format (if not already done)
df_lead['qualified_date'] = pd.to_datetime(df_lead['qualified_date'], errors='coerce')

# Create 'is_qualified' column based on the presence of 'qualified_date'
df_lead['is_qualified'] = df_lead['qualified_date'].notna().astype(int)
```

```python
# Ensure 'days_diff' is not zero to avoid division by zero errors; set to NaN where days_diff <= 0
df_lead['adjusted_days_diff'] = df_lead['days_diff'].apply(lambda x: np.nan if x <= 0 else x)

# Calculate the percentage of total_notes to days_diff
df_lead['notes_days_percentage'] = (df_lead['total_notes'] / df_lead['adjusted_days_diff']) * 100

# Define the follow-up violation based on whether the notes_days_percentage is greater than 50%
# Note: If adjusted_days_diff is NaN (implying original days_diff was 0 or negative), treat these cases separately if needed
df_lead['follow_up_violation'] = df_lead['notes_days_percentage'].apply(lambda x: 1 if x < 50 else 0)
```

```python
import pandas as pd

# Assuming df_lead already contains the 'time_to_first_notes_bucket' and 'is_qualified' columns

# Calculate total leads and qualified leads by time to first notes bucket
pivot_df = pd.pivot_table(filtered_df,
                          index='time_to_first_notes_bucket',
                          values='lead_id',
                          aggfunc='count',
                          margins=True,
                          margins_name='Grand Total')

# Calculate the number of qualified leads by time to first notes bucket
qualified_leads = filtered_df[filtered_df['is_qualified'] == 1].groupby('time_to_first_notes_bucket')['lead_id'].count()

# Add the qualified leads count to the pivot table
pivot_df['Qualified Leads'] = qualified_leads
pivot_df['Qualified Leads'].fillna(0, inplace=True)  # Fill NaN with 0

# Calculate the percentages
pivot_df['% Distribution'] = (pivot_df['lead_id'] / pivot_df.loc['Grand Total', 'lead_id']) * 100
pivot_df['Qualified %'] = (pivot_df['Qualified Leads'] / pivot_df['lead_id']) * 100

# Reorder and rename the columns for clarity
pivot_df.rename(columns={'lead_id': 'Leads'}, inplace=True)
pivot_df = pivot_df[['Leads', '% Distribution', 'Qualified Leads', 'Qualified %']]

# Fill NaN values in the final pivot table, if any, with appropriate values (e.g., 0 or "-")
pivot_df.fillna(0, inplace=True)
```
